5. Calculate the proportion of missing values in each column (use the isnull and sum methods).

```python
data_file.isnull().sum() / len(data_file)
```

```python
for column in data_file.columns:
    missing_percentage = data_file[column].isnull().sum() / data_file.shape[0]
    if missing_percentage:
      print(f"Column '{column}': {missing_percentage:.1%}")
```

6. Remove all columns with missing values except the "Programming Language" column.

```python
cols_to_drop = [col for col in data_file.columns if data_file[col].isnull().sum() > 0 and col != "Язык.программирования"]
data_file.drop(columns=cols_to_drop, inplace=True)
data_file.head(5)
```

7. Calculate again the proportion of missing values in each column and make sure only the "Programming Language" column is left.

```python
data_file.isnull().sum() / len(data_file)
```

```python
for column in data_file.columns:
    missing_percentage = data_file[column].isnull().sum() / data_file.shape[0]
    if missing_percentage:
      print(f"Column '{column}': {missing_percentage:.1%}")
```

8. Remove all rows with missing values in the original table using the dropna method.

```python
data_no_missing = data_file.dropna()
```

9. Determine the new size of the table using the shape method.


```python
rows, columns = data_no_missing.shape
print("Rows:", rows)
print("Columns:", columns)
```

10. Create a new table named python_data containing only rows with specialists who specified Python as their programming language.

```python
python_data = data_file[data_file['Язык.программирования'] == 'Python']
```

11. Determine the size of the python_data table using the shape method.

```python
rows, columns = python_data.shape
print("Rows:", rows)
print("Columns:", columns)
```

12. Use the groupby method to group by the "Position" column.

```python
python_data.groupby(["Должность"])
python_data.head(5)
```

13. Create a new DataFrame where, for the grouped data by "Position" column, perform data aggregation using the agg method and find the minimum and maximum values in the "Salary per Month" column.

```python
grouped_data = data_file.groupby('Должность').agg({'Зарплата.в.месяц': ['min', 'max']})
grouped_data
```

14. Create a function fill_avg_salary that returns the average monthly salary value. Use it with the apply method and create a new column named "avg."

```python
def fill_avg_salary(group):
    avg_salary = group.mean()
    return avg_salary

grouped_data['avg'] = grouped_data.apply(fill_avg_salary, axis=1)
```

15. Generate descriptive statistics using the describe method for the new column.

```python
grouped_data['avg'].describe()
```

16. Save the obtained table to a CSV file.

```python
grouped_data.to_csv('grouped_data.csv')
```

```python
mean_salary_by_city = data_file.groupby('Возраст')['Зарплата.в.месяц'].mean()
fig, ax = plt.subplots(figsize=(12, 6))
di = sns.barplot(x=mean_salary_by_city.index, y=mean_salary_by_city.values, ax=ax)
plt.title('Salary vs. Age')
plt.xlabel('Age')
plt.ylabel('Average salary')
plt.grid()
plt.show()
```

```python
plt.figure(figsize=(15, 6))
sns.scatterplot(x="Возраст", y="Зарплата.в.месяц", data=data_file, color='green')
plt.title('Salary vs. Age')
plt.grid(axis='x')
plt.show()
```

```python
fig, ax = plt.subplots(figsize=(10, 10))
sns.boxenplot(x="Зарплата.в.месяц", y="Должность", data=data_file)
plt.title('Salary Dependence on Position')
plt.grid(axis='y')
plt.show()
```

```python
sns.set(style="whitegrid")
plt.figure(figsize=(15, 10))
ax = sns.boxplot(x="Город", y="Зарплата.в.месяц", data=data_file)
ax.set_xticklabels(ax.get_xticklabels(), rotation=40, ha="right")
ax.set_xlabel('City')
ax.set_ylabel('Monthly Salary')
ax.set_title('Distribution of Salaries by Cities')
sns.despine(trim=True)
plt.show()
```

```python
import plotly.express as px
data_file['Возраст'] = pd.to_numeric(data_file['Возраст'])
mean_salary_by_age = data_file.groupby('Возраст')['Зарплата.в.месяц'].mean().reset_index()
fig = px.bar(mean_salary_by_age, x='Возраст', y='Зарплата.в.месяц', labels={'Возраст': 'Age', 'Зарплата.в.месяц': 'Average Salary'},
             title='Salary vs. Age', template='plotly', width=1000, height=600)
fig.update_layout(title_text='Salary vs. Age', title_x=0.5)
fig.show()
```
