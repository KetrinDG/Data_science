```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
url = "https://uk.wikipedia.org/wiki/%D0%9D%D0%B0%D1%81%D0%B5%D0%BB%D0%B5%D0%BD%D0%BD%D1%8F_%D0%A3%D0%BA%D1%80%D0%B0%D1%97%D0%BD%D0%B8#%D0%9D%D0%B0%D1%80%D0%BE%D0%B4%D0%B6%D1%83%D0%B2%D0%B0%D0%BD%D1%96%D1%81%D1%82%D1%8C"
new_tab = pd.read_html(url, match='Коефіцієнт народжуваності в регіонах України', thousands=".", decimal=",")[0]
```

1. Display the first rows of the table using the head method.

```python
new_tab.head(5)
```

2. Determine the number of rows and columns in the dataframe (shape attribute).

```python
rows, columns = new_tab.shape
print("Rows:", rows)
print("Columns:", columns)
```

3. Replace the values "—" in the table with NaN.

```python
new_tab.replace("—", np.nan, inplace=True)
print(new_tab)
```

4. Determine the data types of all columns using dataframe.dtypes.

```python
new_tab.dtypes
```

5. Convert non-numeric column types to numeric. Hint: These are the columns where the symbol "—" was present.


```python
for column in new_tab.columns:
    if new_tab[column].isnull().any():
        new_tab[column] = new_tab[column].astype(float)
new_tab
```

6. Calculate the proportion of missing values in each column (use the isnull and sum methods).

```python
new_tab.isnull().sum() / len(new_tab)
```

```python
for column in new_tab.columns:
    missing_percentage = new_tab[column].isnull().sum() / new_tab.shape[0]
    if missing_percentage:
      print(f"Column '{column}': {missing_percentage:.1%}")
```

7. Delete data for the entire country, the last row of the table.

```python
new_tab.drop(new_tab.index[-1], inplace=True)
new_tab
```

8. Replace missing data in columns with the column means (fillna method).

```python
new_tab.fillna(new_tab.mean(numeric_only=True), axis=0)
```

9. Obtain a list of regions where the birth rate was higher than the average for Ukraine in 2019.

```python
avg_birth_rate_2019 = new_tab['2019'].mean()
higher_than_avg = new_tab[new_tab['2019'] > avg_birth_rate_2019]['Регіон'].tolist()
print(higher_than_avg)
```

10. Which region had the highest birth rate in 2014?

```python
region_highest_birthrate_2014 = new_tab[new_tab["2014"] == new_tab["2014"].max()]["Регіон"].values[0]
print("Region with Highest Birth Rate in 2014:", region_highest_birthrate_2014)
```

11. Create a bar chart of birth rates by regions in 2019

```python
plt.figure(figsize=(10, 6))
plt.bar(new_tab["Регіон"], new_tab["2019"], color=["b", "r", "y", "g"], edgecolor="black")
plt.xticks(rotation=90)
plt.xlabel("Region", fontsize="small", color="midnightblue")
plt.ylabel("Birth", fontsize="small", color="midnightblue")
plt.title("Chart of birth rates in Ukraine in 2019")
plt.grid(axis='y')
plt.tight_layout()
plt.show()
```

```python
fig, axs = plt.subplots(figsize=(10, 6))
axs.plot(new_tab['Регіон'], new_tab['2019'], marker='o', label='2019')
axs.plot(new_tab['Регіон'], new_tab['2014'], marker='*', label='2014')
plt.plot(new_tab['Регіон'], new_tab['2012'], marker='.', label='2012', linestyle='--')
plt.plot(new_tab['Регіон'], new_tab['2000'], marker='.', label='2000', linestyle='-.')
plt.xlabel('Region', fontsize="small", color="midnightblue")
plt.ylabel('Birth rate', fontsize="small", color="midnightblue")
plt.title('Birth rate by regions of Ukraine')
plt.xticks(rotation=90)
plt.grid()
plt.legend()
plt.show()
```

```python
new_tab_filtered = pd.DataFrame({'Регіон': new_tab['Регіон'], '2000': new_tab['2000']})
new_tab_filtered = new_tab_filtered.set_index('Регіон')
mean_2000 = new_tab_filtered['2000'].mean()
filtered_data = new_tab_filtered[new_tab_filtered['2000'] >= mean_2000]
sorted_data = filtered_data.sort_values(by='2000', ascending=False)
explode = [0.1] * len(sorted_data)
autopct_format = "%.2f%%"
plot = sorted_data.plot.pie(
    y='2000',
    figsize=(10, 10),
    legend=True,
    shadow=True,
    explode=explode,
    autopct=autopct_format,
    pctdistance=1.15,
    labeldistance=1.25,
)
plot.legend(bbox_to_anchor=(1, 1.02), loc='best')
plt.figtext(0.5, 0.95, 'Birth rate by regions of Ukraine in 2000', ha='center', fontsize=16, color="blue")
plt.show()
```
