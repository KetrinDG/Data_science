Loading data from the 'data_2d.csv' file and standardization using StandardScaler.

```python
df_2d = pd.read_csv('data_2d.csv')
df_2d_scaler = StandardScaler().fit_transform(df_2d.values)
```

Loading data from the 'mnist.csv' file and standardizing it using StandardScaler.

```python
df_mnist = pd.read_csv('mnist.csv')
df_mnist_scaler = StandardScaler().fit_transform(df_mnist.values[:, 1:])
```

Calculating the Sum of Squared Errors (SSE) for different numbers of clusters.

```python
sse = []
for k in range(2, 11):
    kmeans = KMeans(n_clusters=k, random_state=11)
    kmeans.fit(df_2d_scaler)
    sse.append(kmeans.inertia_)
```

Plotting the dependency graph.

```python
plt.plot(range(2, 11), sse, marker='o')
plt.title('Elbow Method for Optimal Cluster Selection')
plt.xlabel('Number of clusters')
plt.ylabel('SSE')
plt.show()
```

Clustering data using the optimal number of clusters (optimal_k_2d).

```python
optimal_k_2d = 4
kmeans = KMeans(n_clusters=optimal_k_2d, random_state=11)
kmeans.fit(df_2d_scaler)
labels = kmeans.labels_
```

Plotting the dependency graph.

```python
plt.scatter(df_2d.values[:, 0], df_2d.values[:, 1], c=labels)
plt.title('K-means 2d Dataset')
plt.xlabel('First principal component')
plt.ylabel('Second principal component')
plt.show()
```

Calculating the Sum of Squared Errors (SSE) for different numbers of clusters in the MNIST dataset.

```python
see_mnist = []
for k in range(2, 11):
    kmeans = KMeans(n_clusters=k, random_state=11)
    kmeans.fit(df_mnist_scaler)
    see_mnist.append(kmeans.inertia_)
```

Plotting the dependency graph.

```python
plt.plot(range(2, 11), see_mnist, marker='o')
plt.title('Elbow Method for Optimal Cluster Selection')
plt.xlabel('Number of clusters')
plt.ylabel('SSE')
plt.show()
```


Clustering MNIST data with the optimal number of clusters.

```python
optimal_k_mnist = 6
kmeans = KMeans(n_clusters=optimal_k_mnist, random_state=11)
kmeans.fit(df_mnist_scaler)
labels_mnist = kmeans.labels_
```


Applying Principal Component Analysis (PCA) to reduce the dimensionality of MNIST data to 2 components.

```python
pca = PCA(n_components=2)
reduced_data = pca.fit_transform(df_mnist_scaler)
```

Plotting the dependency graph.

```python
plt.scatter(reduced_data[:, 0], reduced_data[:, 1], c=labels_mnist)
plt.title('K-means Clustering (MNIST Dataset)')
plt.xlabel('First principal component')
plt.ylabel('Second principal component')
plt.show()
```
